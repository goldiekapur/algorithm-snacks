```python
import cv2
import pandas as pd
from PIL import Image
import os
import numpy as np
from matplotlib.pyplot import imshow
```

```python
xml_cvs = '/home/wengzh/image_recognition/image_data/food_od_all/all_xml.csv'
full_labels = pd.read_csv(xml_cvs)
```

```python
full_labels.head()
```

```python
def draw_boxes(image_dir, image_name):
    '''
    - image_dir: dir contains all images
    - image_name: elements in full_labels.filename
    '''
    sel_value = full_labels[full_labels.filename == image_name]
    img = cv2.imread(os.path.join(image_dir, image_name))
    for index, row in sel_value.iterrows():
        img = cv2.rectangle(img, (row['xmin'], row['ymin']), (row['xmax'], row['ymax']), (0, 255, 0), 3)
    return img
```

```python
dir_path = '/home/wengzh/image_recognition/image_data/food_od_all/check'
Image.fromarray(draw_boxes(dir_path,'food_1_52.jpg'))
```
